```python
import numpy as np
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=2)
gmm.fit(df_scaled)

#predictions from gmm
labels = gmm.predict(df_scaled)
frame = pd.DataFrame(df_scaled)
frame['cluster'] = labels

frame.columns = ['IsCanceled','LeadTime',  'ArrivalDateWeekNumber','StaysInWeekNights', 
                    'ADR','TotalOfSpecialRequests','cluster']

color=['blue','green','cyan', 'black']
for k in range(0,4):
    data = frame[frame["cluster"]==k]
    plt.scatter(data["ArrivalDateWeekNumber"],data["LeadTime"],c=color[k])
plt.show()
```

```python
from sklearn.metrics.cluster import adjusted_rand_score
from sklearn.cluster import DBSCAN
import mglearn
y=X['IsCanceled']
fig, axes = plt.subplots(1, 4, figsize=(15, 3),
                         subplot_kw={'xticks': (), 'yticks': ()})

# make a list of algorithms to use
algorithms = [AgglomerativeClustering(n_clusters=2), GaussianMixture(n_components=2),
              DBSCAN()]

for ax, algorithm in zip(axes[1:], algorithms):
    clusters = algorithm.fit_predict(df_scaled)
    ax.scatter(df_scaled[:, 2], df_scaled[:, 1], c=clusters,
               cmap=mglearn.cm3, s=60)
    ax.set_title("{} - ARI: {:.2f}".format(algorithm.__class__.__name__,
                                           adjusted_rand_score(y, clusters)))
```

```python
from sklearn.decomposition import PCA
sk_pca = PCA(n_components=6,random_state=0 )
scaled=scale(Xx)

pca_op = sk_pca.fit(scaled)
pca_t = sk_pca.transform(scaled)
```

```python
dset2 = pd.DataFrame()
dset2['pca'] = range(1,7)
dset2['vari'] = pd.DataFrame(pca_op.explained_variance_ratio_)
plt.figure(figsize=(8,6))
graph = sns.barplot(x='pca', y='vari', data=dset2)
for p in graph.patches:
    graph.annotate('{:.2f}'.format(p.get_height()), (p.get_x()+0.2, p.get_height()),
                   ha='center', va='bottom',
                   color= 'black')
plt.ylabel('Proportion', fontsize=18)
plt.xlabel('Principal Component', fontsize=18)
plt.show() 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(pca_t,y,test_size=0.3,random_state=42)
```

```python
dt_reg = DecisionTreeRegressor()
dt_reg.fit(X_train, y_train)
```

```python
print("Accuracy of test is:", dt_reg.score(X_test, y_test))
```

```python
dt_reg.feature_importances_
```

```python
ridge_grid = {'alpha':[0.00001, 0.0001, 0.001, 0.01, 0.1, 1, 10,100,1000,10000]}
lasso=Lasso()
lasso_cv_reg=GridSearchCV(lasso,ridge_grid, cv=5)
lasso_cv_reg.fit(X_train,y_train)
```

```python
print('Mean validaton score is:', lasso_cv_reg.best_score_)
#Performance measure of Lasso model(Using R2)
print('Unbiased Performance of best model:',lasso_cv_reg.score(X_test,y_test))
# Finding the best hyperparameter
print('Optimal Hyperparam:',lasso_cv_reg.best_params_)
# Finding the best parameter estimates
lasso_cv_reg.best_estimator_.coef_,lasso_cv_reg.best_estimator_.intercept_
```

```python
ridge_grid = {'alpha':[0.001, 0.01, 0.1,1, 10,100]} 

ridge=Ridge()
ridge_cv_reg=GridSearchCV(ridge,ridge_grid, cv=5) #experimented cv's are 4,5,8,9
ridge_cv_reg.fit(X_train,y_train)
```
